### np.zeros() np.ones()
- To intialize the weights during the first iteration of Tensorflow and statistics flow

```python
x = np.zeros((2,2)) # datatype as float
print(x)
```

```python
# 3 to represent the values as interger
x = np.zeros((2,2),dtype=np.int16)
print(x)
```

### Vizualizing the data
- this is an interesting way to display Dataframe

- with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
divisions = [1,2,3,4]
marks = [79,82,75,63]
width =0.70
plt.bar(divisions,marks,color='blue',width=0.3)
plt.show()
```

```python
plt.barh(divisions,marks,color='blue')
plt.show()
```

```python
lang = ['Python','Java','c','c++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
